# JSON (JavaScript Object Notation) - Example

JSON is a strict data format built for machines to read and write. Everything is wrapped in curly braces `{}` for objects and square brackets `[]` for arrays, and all keys must be double-quoted strings. It's the standard format for APIs and web services — when your app talks to a server, the response is almost always JSON. No comments allowed, and every comma and bracket must be exactly right or the file breaks.

````python
import json

with open('example.json', 'r') as f:
    profile_json = json.load(f)

print(type(profile_json))
print(profile_json)
````

````python
print('Name:  ', profile_json['name'])
print('Role:  ', profile_json['role'])
print('Active:', profile_json['active'])
print('Notes: ', profile_json['notes'])
````

````python
contact = profile_json['contact']
print('Email:   ', contact['email'])
print('GitHub:  ', contact['github'])

print('\nSkills:')
for skill in profile_json['skills']:
    print(' -', skill)
````

````python
print('Courses:')
for course in profile_json['courses']:
    status = '✅' if course['completed'] else '🔲'
    score  = course['score'] if course['score'] is not None else 'N/A'
    print(f"  {status} {course['title']} ({course['level']}) — Score: {score}")
````

````python
profile_json['age'] = 29
profile_json['skills'].append('Matplotlib')

with open('developer_profile_updated.json', 'w') as f:
    json.dump(profile_json, f, indent=2)

print('Saved updated JSON.')
````

# YAML (YAML Ain't Markup Language) - Example

YAML is JSON's more human-friendly cousin. It represents the exact same data structures but ditches most of the punctuation — no curly braces, no quotes around keys, just clean indentation and colons. It's the go-to format for configuration files (Docker Compose, GitHub Actions, Kubernetes). The tradeoff: whitespace *is* syntax, so a misplaced indent can silently break things. Supports `#` comments, which JSON does not.

````python
import yaml

with open('example.yaml', 'r') as f:
    profile_yaml = yaml.safe_load(f)

print(type(profile_yaml))
print(profile_yaml)
````

````python
print('Name:  ', profile_yaml['name'])
print('Role:  ', profile_yaml['role'])
print('Notes: ', profile_yaml['notes'])
print('Email: ', profile_yaml['contact']['email'])
print('Skills:', profile_yaml['skills'])
````

````python
with open('example.json') as f:
    profile_json_clean = json.load(f)

print('JSON == YAML:', profile_json_clean == profile_yaml)
````

````python
multi_yaml = """
name: Alice
role: Data Scientist
---
name: Bob
role: ML Engineer
"""

docs = list(yaml.safe_load_all(multi_yaml))
for doc in docs:
    print(doc)
````

````python
profile_yaml['age'] = 29

with open('developer_profile_updated.yaml', 'w') as f:
    yaml.dump(profile_yaml, f, default_flow_style=False, sort_keys=False)

print('Saved updated YAML.')
````

# Markdown Example

Markdown is not a data format — it's a *document* format. Where JSON and YAML store structured data for programs to consume, Markdown stores formatted text for humans to read. You write plain text with lightweight symbols (`#` for headings, `**bold**`, `- lists`) and it renders into styled HTML. It's used everywhere developers write: READMEs, documentation, Jupyter notebooks, and blog posts. It has no concept of data types or key-value pairs.

````python
with open('example.md', 'r') as f:
    md_text = f.read()

print(md_text[:300])
````

````python
import re

headings = re.findall(r'^(#{1,6})\s+(.+)', md_text, re.MULTILINE)

print('Headings found:')
for hashes, title in headings:
    level = len(hashes)
    indent = '  ' * (level - 1)
    print(f"{indent}H{level}: {title}")
````

````python
def get_section(md: str, heading: str) -> str:
    """Return the text content under a given ## heading."""
    pattern = rf'(?<=## {re.escape(heading)}\n)(.*?)(?=\n## |\Z)'
    match = re.search(pattern, md, re.DOTALL)
    return match.group(0).strip() if match else 'Section not found'

print('=== Skills section ===')
print(get_section(md_text, 'Skills'))

print('\n=== Contact section ===')
print(get_section(md_text, 'Contact'))
````

````python
links = re.findall(r'\[([^\]]+)\]\(([^)]+)\)', md_text)

print('Links found:')
for text, url in links:
    print(f'  Text: "{text}"  →  URL: {url}')
````

````python
code_blocks = re.findall(r'```(?:\w+)?\n(.*?)```', md_text, re.DOTALL)

print(f'Found {len(code_blocks)} code block(s):\n')
for i, block in enumerate(code_blocks, 1):
    print(f'--- Block {i} ---')
    print(block.strip())
````
